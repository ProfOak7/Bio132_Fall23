```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
from scipy.stats import sem
from statistics import mean, stdev
```

```python
#Optional for disabling warnings 
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
#Set figure resolution to 300dpi
import matplotlib as mpl
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
df = pd.read_excel (r'./ZOI_compounds.xlsx') 
df
```

```python
#To make the dataset the same shape I'm going to drop rows that contain NaN values
df = df.dropna()
```

```python
AMOX = df.Amox
APA = df.APA
A = df.A
B = df.B
C1 = df.C1
C2 = df.C2
D = df.D
E = df.E
F = df.F
G = df.G1
H = df.H
I = df.I
J = df.J
H2O = df.H2O
```

```python
AMOX_mean = mean(AMOX)
APA_mean = mean(APA)
A_mean = mean(A)
B_mean = mean(B)
C1_mean = mean(C1)
C2_mean = mean(C2)
D_mean= mean(D)
E_mean = mean(E)
F_mean = mean(F)
G_mean = mean(G)
H_mean = mean(H)
I_mean = mean(I)
J_mean = mean(J)
H2O_mean= mean(H2O)
```

```python
AMOX_stdev = stdev(AMOX)
APA_stdev = stdev(APA)
A_stdev = stdev(A)
B_stdev = stdev(B)
C1_stdev = stdev(C1)
C2_stdev = stdev(C2)
D_stdev= stdev(D)
E_stdev = stdev(E)
F_stdev = stdev(F)
G_stdev = stdev(G)
H_stdev = stdev(H)
I_stdev = stdev(I)
J_stdev = stdev(J)
H2O_stdev= stdev(H2O)
```

```python
AMOX_sem = sem(AMOX)
APA_sem = sem(APA)
A_sem = sem(A)
B_sem = sem(B)
C1_sem = sem(C1)
C2_sem = sem(C2)
D_sem= sem(D)
E_sem = sem(E)
F_sem = sem(F)
G_sem = sem(G)
H_sem = sem(H)
I_sem = sem(I)
J_sem = sem(J)
H2O_sem= sem(H2O)
```

```python
# build a bar plot
plt.bar(['Amox', '6-APA','A','B','C1','C2','D','E','F','G','H','I','J','H2O'],[AMOX_mean, APA_mean,A_mean,B_mean,C1_mean,C2_mean,D_mean,E_mean,F_mean,G_mean,H_mean,I_mean,J_mean, H2O_mean], 0.7,  yerr=[AMOX_sem, APA_sem,A_sem,B_sem,C1_sem,C2_sem,D_sem,E_sem,F_sem,G_sem,H_sem,I_sem,J_sem, H2O_sem], capsize=10, edgecolor='black')
plt.xticks(rotation=55)
plt.ylabel('Zone of Inhibition (mm)')
plt.xlabel("Compound")
plt.title('Inhibition of Compounds on Bacterial Growth (E. coli)')
```
