```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
from scipy.stats import sem
from statistics import mean, stdev
```

```python
#Optional for disabling warnings 
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
#Set figure resolution to 300dpi
import matplotlib as mpl
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
df = pd.read_excel (r'./ZOI_practice.xlsx') 
df
```

```python
ecoli_half = df.half
ecoli_third = df.third
ecoli_eleventh = df.eleventh
ecoli_zero = df.zero
sepi_half = df.halfs
sepi_third = df.thirds
sepi_eleventh = df.elevenths
sepi_zero = df.zeros

```

```python
ECOLI_HALF_MEAN = mean(ecoli_half)
ECOLI_HALF_STDEV = stdev(ecoli_half)
ECOLI_HALF_STERROR = sem(ecoli_half)
ECOLI_THIRD_MEAN = mean(ecoli_third)
ECOLI_THIRD_STDEV = stdev(ecoli_third)
ECOLI_THIRD_STERROR = sem(ecoli_third)
ECOLI_EL_MEAN = mean(ecoli_eleventh)
ECOLI_EL_STDEV = stdev(ecoli_eleventh)
ECOLI_EL_STERROR = sem(ecoli_eleventh)
ECOLI_ZERO_MEAN = mean(ecoli_zero)
ECOLI_ZERO_STDEV = stdev(ecoli_zero)
ECOLI_ZERO_STERROR = sem(ecoli_zero)

```

```python
# build a bar plot
plt.bar(['1:1', '1:2','1:10','0'],[ECOLI_HALF_MEAN, ECOLI_THIRD_MEAN, ECOLI_EL_MEAN, ECOLI_ZERO_MEAN], yerr=[ECOLI_HALF_STERROR, ECOLI_THIRD_STERROR, ECOLI_EL_STERROR, ECOLI_ZERO_STERROR], capsize=10)
plt.ylabel('Zone of Inhibition (mm)')
plt.title('Inhibition of Ampicillin on Bacterial Growth (e. coli)')
```

```python
SEPI_HALF_MEAN = mean(sepi_half)
SEPI_HALF_STDEV = stdev(sepi_half)
SEPI_HALF_STERROR = sem(sepi_half)
SEPI_THIRD_MEAN = mean(sepi_third)
SEPI_THIRD_STDEV = stdev(sepi_third)
SEPI_THIRD_STERROR = sem(sepi_third)
SEPI_EL_MEAN = mean(sepi_eleventh)
SEPI_EL_STDEV = stdev(sepi_eleventh)
SEPI_EL_STERROR = sem(sepi_eleventh)
SEPI_ZERO_MEAN = mean(sepi_zero)
SEPI_ZERO_STDEV = stdev(sepi_zero)
SEPI_ZERO_STERROR = sem(sepi_zero)
```

```python
# build a bar plot
plt.bar(['1:1', '1:2','1:10','0'],[SEPI_HALF_MEAN, SEPI_THIRD_MEAN, SEPI_EL_MEAN, SEPI_ZERO_MEAN], yerr=[SEPI_HALF_STERROR, SEPI_THIRD_STERROR, SEPI_EL_STERROR, SEPI_ZERO_STERROR], capsize=10)
plt.ylabel('Zone of Inhibition (mm)')
plt.title('Inhibition of Ampicillin on Bacterial Growth (s. epidermis)')
```

```python
X = ['1:1', '1:2','1:10','0']
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2,[ECOLI_HALF_MEAN, ECOLI_THIRD_MEAN, ECOLI_EL_MEAN, ECOLI_ZERO_MEAN], 0.4, yerr=[ECOLI_HALF_STERROR, ECOLI_THIRD_STERROR, ECOLI_EL_STERROR, ECOLI_ZERO_STERROR], label = 'e. coli', capsize=10)
plt.bar(X_axis + 0.2,[SEPI_HALF_MEAN, SEPI_THIRD_MEAN, SEPI_EL_MEAN, SEPI_ZERO_MEAN], 0.4, yerr=[SEPI_HALF_STERROR, SEPI_THIRD_STERROR, SEPI_EL_STERROR, SEPI_ZERO_STERROR], label = 's. epidermis', capsize=10)
plt.xticks(X_axis, X)
plt.ylabel('Zone of Inhibition (mm)')
plt.title('Inhibition of Ampicillin on Bacterial Growth')
plt.legend()
plt.show()
```
